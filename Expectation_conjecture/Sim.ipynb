```python
import pandas as pd
import matplotlib.pyplot as plt

def combine_n_csv_files(file_paths, output_file):
    
    dataframes = []
    
    for file_path in file_paths:
        print(f"Processing file: {file_path}")
        df = pd.read_csv(file_path)
        dataframes.append(df)
    
    combined_df = pd.concat(dataframes, ignore_index=True)
    
    combined_df.to_csv(output_file, index=False)
    print(f"Combined CSV saved to: {output_file}")

file_paths = [
    "random_combinations_results_T1.csv",  
    "random_combinations_results_T2.csv",
    "random_combinations_results_T3.csv",
    "random_combinations_results_T4.csv",
    "random_combinations_results_T5.csv",
    "random_combinations_results_T6.csv",
    "random_combinations_results_T7.csv",
    "random_combinations_results_T8.csv",
    "random_combinations_results_T9.csv",
]

file_paths_zoomed = [
    "random_combinations_results_T7.csv",
    "random_combinations_results_T8.csv",
    "random_combinations_results_T9.csv",
]

output_file = "final_all.csv"
output_file_zoomed = "final_all_zoomed.csv"

# Combine the 9 files
combine_n_csv_files(file_paths, output_file)

combine_n_csv_files(file_paths_zoomed, output_file_zoomed)
```

```python
file_path = "final_all.csv" 
data = pd.read_csv(file_path)

data['Expectation'] = data['Expectation'].apply(eval)

data['Positive_Comparison'] = data['T'] / (2 ** (data['T'] + 1))
data['Negative_Comparison'] = -data['T'] / (2 ** (data['T'] + 1))

data = data.sort_values(by="T")

pastel_blue = "#92c5de"
pastel_red = "#f4a582"
pastel_green = "#b8e186"

plt.figure(figsize=(12, 8))
plt.scatter(data['T'], data['Expectation'], color=pastel_blue, alpha=0.7, label="Expectation values")
plt.plot(data['T'], data['Positive_Comparison'], color=pastel_green, linestyle='-', marker='o', linewidth=2, markersize=6, label="T / 2^(T+1)")
plt.plot(data['T'], data['Negative_Comparison'], color=pastel_red, linestyle='-', marker='o', linewidth=2, markersize=6, label="-T / 2^(T+1)")
plt.axhline(0, color='gray', linestyle='-', linewidth=1, alpha=0.6, label="Zero Reference")
plt.grid(color='gray', linestyle='-', linewidth=0.5, alpha=0.5)
plt.xlabel("T", fontsize=14)
plt.ylabel("Values", fontsize=14)
plt.title("Expectation Values", fontsize=16)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```
